```python
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=dataset)
sns.heatmap(dataset.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
```

```python
sns.histplot(dataset['Married'])
sns.boxplot(x=dataset['Loan_Status'])
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
status_counts = dataset['Loan_Status'].value_counts()
status_percent = dataset['Loan_Status'].value_counts(normalize=True) * 100
sns.barplot(x=status_counts.index, y=status_percent.values, palette='Set2')
plt.title('Loan Status Percentage')
plt.ylabel('Percentage')
plt.xlabel('Loan Status')
```

```python
plt.subplot(1, 3, 2)
married_status = pd.crosstab(dataset['Married'], dataset['Loan_Status'], normalize='index') * 100
married_status.plot(kind='bar', stacked=True, ax=plt.gca(), colormap='Pastel1')
plt.title('Loan Status by Marital Status')
plt.ylabel('Percentage')
```

```python
dataset['Total_Income'] = dataset['ApplicantIncome'] + dataset['CoapplicantIncome']
plt.subplot(1, 3, 3)
sns.boxplot(x='Loan_Status', y='Total_Income', data=dataset, palette='coolwarm')
plt.title('Total Income vs Loan Status')
plt.ylabel('Total Income')
plt.xlabel('Loan Status')

plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', hue='Loan_Status', data=dataset)
plt.title('Applicant Income vs Loan Amount')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```

```python
loan_status_counts = dataset['Loan_Status'].value_counts()
plt.pie(loan_status_counts, labels=loan_status_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral'])
plt.title('Loan Status Distribution')
plt.axis('equal')  # Make it a circle
plt.show()
```

```python
sns.violinplot(x='Education', y='LoanAmount', data=dataset, palette='coolwarm')
plt.title('Loan Amount Distribution by Education')
plt.show()
```

```python
# Clean the dataset: drop non-numeric
numeric_dataset = dataset.select_dtypes(include='number')

# Correlation matrix + heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_dataset.corr(), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```
